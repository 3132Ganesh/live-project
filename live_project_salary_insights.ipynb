#### 3.1. Initializing Data for Cleaning

```python
import numpy as np
import pandas as pd

def clean_ctc(val):
  if pd.isna(val):
    return np.nan
  val_str = str(val).strip().lower()
  # Remove 'lpa', ',', '₹' to get a clean numeric string
  val_str = val_str.replace('lpa', '').replace(',', '').replace('₹', '').strip()

  if not val_str:
    return np.nan

  try:
    numeric_val = float(val_str)
    # Heuristic: If the numeric value is 1000 or more, assume it's an absolute value in rupees
    # that needs to be converted to LPA by dividing by 100,000.
    # Otherwise, assume it's already in LPA.
    if numeric_val >= 1000:
      return numeric_val / 100000
    else: # Assume it's already in LPA
      return numeric_val
  except ValueError:
    return np.nan
```

### Final Report: Bangalore Tech Salaries Analysis

```python
# Q3.1 — CTC distribution by role
role_ctc_distribution = dd.groupby('role_clean')['current_ctc'].agg(['median', 'mean', 'min', 'max']).sort_values(by='median', ascending=False)
```

```python
# Q3.2 — Experience curve for SDE Backend
sde_backend_df = dd[dd['role_clean'] == 'SDE Backend'].copy()
sde_backend_df['years_exp_band'] = pd.cut(sde_backend_df['years_exp'],
                                          bins=[0, 1, 3, 5, np.inf],
                                          labels=['0-1', '2-3', '4-5', '6+'],
                                          right=False)
sde_exp_curve = sde_backend_df.groupby('years_exp_band')['current_ctc'].median().reset_index()
sde_exp_curve['ctc_growth'] = sde_exp_curve['current_ctc'].diff().fillna(0)
```

```python
# Q3.3 — Skill premium for SDEs
sde_roles = ['SDE Backend', 'SDE Frontend', 'SDE Fullstack']
sde_df = dd[dd['role_clean'].isin(sde_roles)].copy()

selected_skills = ['AWS', 'ML', 'System Design', 'Kubernetes']
skill_premiums = []

for skill in selected_skills:
    has_skill_df = sde_df[sde_df['skills'].fillna('').str.contains(skill, case=False, na=False)]
    no_skill_df = sde_df[~sde_df['skills'].fillna('').str.contains(skill, case=False, na=False)]

    median_has = has_skill_df['current_ctc'].median()
    median_no = no_skill_df['current_ctc'].median()

    skill_premiums.append({
        'skill': skill,
        'median_with_skill': median_has,
        'median_without_skill': median_no,
        'premium': median_has - median_no
    })
skill_premiums_df = pd.DataFrame(skill_premiums).sort_values(by='premium', ascending=False)
```

```python
# Q3.4 — Company-type premium, same role
target_role_q4 = 'SDE Backend'
role_company_ctc = dd[dd['role_clean'] == target_role_q4].groupby('company_type_clean')['current_ctc'].median()

unicorn_ctc_q4 = role_company_ctc.get('Unicorn', np.nan)
mnc_ctc_q4 = role_company_ctc.get('MNC', np.nan)

company_type_premium = np.nan
if not np.isnan(unicorn_ctc_q4) and not np.isnan(mnc_ctc_q4) and mnc_ctc_q4 != 0:
    company_type_premium = ((unicorn_ctc_q4 - mnc_ctc_q4) / mnc_ctc_q4) * 100
```

```python
# Q3.5 — Underpaid professionals

dd['exp_band'] = pd.cut(dd['years_exp'],
                        bins=[0, 1, 3, 5, np.inf],
                        labels=['0-1', '2-3', '4-5', '6+'],
                        right=False)

group_medians = dd.groupby(['role_clean', 'company_type_clean', 'exp_band'])['current_ctc'].transform('median')
dd['group_median_ctc'] = group_medians

# Filter groups with at least 10 members
group_counts = dd.groupby(['role_clean', 'company_type_clean', 'exp_band']).size()
reliable_groups = group_counts[group_counts >= 10].index
dd_reliable = dd[dd.set_index(['role_clean', 'company_type_clean', 'exp_band']).index.isin(reliable_groups)].copy()

dd_reliable['gap'] = dd_reliable['current_ctc'] - dd_reliable['group_median_ctc']

top_underpaid = dd_reliable.sort_values(by='gap').head(10)
```

```python
# --- Format and Print Report --- #

report = []
report.append("=" * 80)
report.append(f"{'Bangalore Tech Salaries Insights':^80}")
report.append("=" * 80)
report.append("\n")

report.append("Q3.1 — CTC Distribution by Role (LPA)")
report.append("-" * 80)
report.append(f"{'Role':<30} {'Median':>10} {'Mean':>10} {'Min':>10} {'Max':>10}")
report.append("-" * 80)
for index, row in role_ctc_distribution.head(5).iterrows(): # Display top 5 for brevity
    report.append(f"{index:<30} {row['median']:>10.2f} {row['mean']:>10.2f} {row['min']:>10.2f} {row['max']:>10.2f}")
report.append("-" * 80)
report.append(f"Most Paid: {role_ctc_distribution.index[0]} | Least Paid: {role_ctc_distribution.index[-1]}")
report.append("\n")

report.append("Q3.2 — SDE Backend Experience Curve (Median CTC in LPA)")
report.append("-" * 80)
report.append(f"{'Exp. Band':<20} {'Median CTC':>15} {'CTC Growth':>15}")
report.append("-" * 80)
for index, row in sde_exp_curve.iterrows():
    report.append(f"{row['years_exp_band']:<20} {row['current_ctc']:>15.2f} {row['ctc_growth']:>15.2f}")
report.append("\n")

report.append("Q3.3 — SDE Skill Premiums (Median CTC in LPA)")
report.append("-" * 80)
report.append(f"{'Skill':<20} {'With Skill':>15} {'Without Skill':>15} {'Premium':>15}")
report.append("-" * 80)
for index, row in skill_premiums_df.iterrows():
    report.append(f"{row['skill']:<20} {row['median_with_skill']:>15.2f} {row['median_without_skill']:>15.2f} {row['premium']:>15.2f}")
report.append("\n")

report.append(f"Q3.4 — Company-Type Premium for {target_role_q4} (Median CTC in LPA)")
report.append("-" * 80)
report.append(f"{'Company Type':<20} {'Median CTC':>15}")
report.append("-" * 80)
for company_type, ctc in role_company_ctc.items():
    report.append(f"{company_type:<20} {ctc:>15.2f}")
report.append("-" * 80)
if not np.isnan(company_type_premium):
    report.append(f"Unicorn premium over MNC: {company_type_premium:.2f}%")
report.append("\n")

report.append("Q3.5 — Top 10 Underpaid Professionals (LPA)")
report.append("-" * 80)
report.append(f"{'ID':<15} {'Role':<25} {'Company Type':<15} {'CTC':>8} {'Median':>8} {'Gap':>8}")
report.append("-" * 80)
for index, row in top_underpaid.iterrows():
    report.append(f"{row['employee_id']:<15} {row['role_clean']:<25} {row['company_type_clean']:<15} {row['current_ctc']:>8.2f} {row['group_median_ctc']:>8.2f} {row['gap']:>8.2f}")
report.append("=" * 80)

print("\n".join(report))
```
